# Generators in Python

## What is a Generator?

A **generator** is a special type of **iterator** that generates values **one at a time** using the `yield` keyword instead of returning all values at once.

A generator follows the **Iterator Protocol**, meaning it automatically implements:

- `__iter__()` → Returns the generator object.
- `__next__()` → Returns the next generated value. Raises `StopIteration` when no values remain.

---

## Why Use Generators?

- Generate values one at a time.
- Save memory using lazy evaluation.
- Process large datasets efficiently.
- Simplify iterator creation.
- Improve performance for data streams.

---

## How a Generator Works

```text
Generator Function
        │
        ▼
Call Function
        │
        ▼
Generator Object
        │
   next(generator)
        │
        ▼
Returns one value
        │
      Pauses
        │
   next(generator)
        │
        ▼
Resumes from previous yield
```

---

## Creating a Generator

```python
def numbers():
    yield 10
    yield 20
    yield 30

gen = numbers()

print(next(gen))
print(next(gen))
print(next(gen))
```

Output

```text
10
20
30
```

---

## How `yield` Works

Unlike `return`, `yield` **does not terminate** the function.

It:

- Returns a value.
- Pauses the function.
- Saves the current state.
- Continues from the same position when `next()` is called again.

Example

```python
def demo():
    print("Start")
    yield 1

    print("Middle")
    yield 2

    print("End")
    yield 3

gen = demo()

print(next(gen))
print(next(gen))
print(next(gen))
```

Output

```text
Start
1
Middle
2
End
3
```

---

## StopIteration

When all values have been generated, Python raises a `StopIteration` exception.

```python
def numbers():
    yield 1
    yield 2

gen = numbers()

print(next(gen))
print(next(gen))
print(next(gen))
```

Output

```text
1
2
StopIteration
```

## Generator with a Loop

```python
def numbers():
    for i in range(5):
        yield i

for value in numbers():
    print(value)
```

Output

```text
0
1
2
3
4
```

---

## Generator vs Function

| Function | Generator |
|----------|-----------|
| Uses `return` | Uses `yield` |
| Returns all data at once | Produces one value at a time |
| Execution ends after `return` | Execution pauses after `yield` |
| Higher memory usage | Lower memory usage |

---

## Iterator vs Generator

| Iterator | Generator |
|----------|-----------|
| Created using `iter()` | Created using `yield` |
| Requires manual implementation of `__iter__()` and `__next__()` | Automatically implements both methods |
| More code | Less code |
| Harder to create | Easier to create |

---

## Real-World Example

Imagine a **coffee machine**.

- You press the button.
- It prepares **one cup** of coffee.
- You press again.
- It prepares the **next cup**.

The machine does **not** prepare all cups at once.

A generator works the same way—it produces **one value at a time** only when requested.

---

## Key Functions

| Function | Purpose |
|----------|---------|
| `yield` | Produces the next value and pauses execution |
| `next()` | Retrieves the next generated value |

---

## Key Takeaways

- A generator is a special type of iterator.
- Generators use the `yield` keyword.
- `yield` pauses execution instead of terminating the function.
- Values are generated one at a time.
- Generators are memory-efficient because they use lazy evaluation.
- Every generator is an iterator, but not every iterator is a generator.
