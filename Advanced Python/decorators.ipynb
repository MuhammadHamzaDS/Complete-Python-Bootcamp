# Decorators

## What is a Decorator?

A **decorator** is a function that **adds extra functionality to another function without modifying its original source code**.

---

## Syntax

```python
def decorator(func):

    def wrapper():
        # Extra functionality
        func()

    return wrapper
```

---

## Using a Decorator

```python
@decorator
def greet():
    print("Hello")

greet()
```

Equivalent to:

```python
def greet():
    print("Hello")

greet = decorator(greet)

greet()
```

---

## How a Decorator Works

1. The original function is passed to the decorator.
2. The decorator creates a `wrapper()` function.
3. The `wrapper()` adds extra functionality.
4. The decorator returns the `wrapper()`.
5. The original function reference is replaced by the `wrapper()`.

```
Original Function
        │
        ▼
Decorator
        │
        ▼
Wrapper Function
        │
        ▼
Returned
        │
        ▼
Original Function = Wrapper
```

---

## Decorator with Arguments

```python
def repeat(n):

    def decorator(func):

        def wrapper(*args, **kwargs):
            for _ in range(n):
                func(*args, **kwargs)

        return wrapper

    return decorator
```

Usage:

```python
@repeat(3)
def hello():
    print("Hello")

hello()
```

Output:

```text
Hello
Hello
Hello
```

Equivalent to:

```python
def hello():
    print("Hello")

hello = repeat(3)(hello)

hello()
```

---

## Why `wrapper()`?

The `wrapper()` function executes the original function and adds extra behavior before or after it.

---

## Why `*args` and `**kwargs`?

They allow the decorator to work with **any function**, regardless of the number or type of arguments.

```python
@repeat(2)
def add(a, b):
    print(a + b)

add(2, 3)
```

Output:

```text
5
5
```

---

## Common Use Cases

* Logging
* Authentication
* Authorization
* Caching
* Execution Timing
* Input Validation

---

## Key Takeaways

* A decorator is a function that takes another function as input.
* It adds extra functionality without changing the original function.
* `wrapper()` contains the additional behavior.
* `@decorator` is syntactic sugar for:

```python
function = decorator(function)
```

* A parameterized decorator works as:

```python
function = repeat(3)(function)
```

* `*args` and `**kwargs` make decorators compatible with any function.
