# Iterators in Python

## What is an Iterator?

An **iterator** is an object that allows you to traverse a collection **one element at a time** without exposing its underlying implementation.

In Python, an iterator follows the **Iterator Protocol**, meaning it implements two special methods:

- `__iter__()` → Returns the iterator object itself.
- `__next__()` → Returns the next element. Raises `StopIteration` when no elements remain.

---

## Why Use Iterators?

- Process data one item at a time.
- Improve memory efficiency.
- Work with large datasets.
- Enable lazy evaluation.

---

## How an Iterator Works

```text
Iterable (list, tuple, string, etc.)
            │
            ▼
      iter(iterable)
            │
            ▼
        Iterator
            │
      next(iterator)
            │
            ▼
Returns one element at a time
```

---

## Creating an Iterator

```python
numbers = [10, 20, 30]

iterator = iter(numbers)

print(next(iterator))   # 10
print(next(iterator))   # 20
print(next(iterator))   # 30
```

Output

```text
10
20
30
```

---

## `StopIteration`

When no elements remain, Python raises a `StopIteration` exception.

```python
numbers = [1, 2]

iterator = iter(numbers)

print(next(iterator))
print(next(iterator))
print(next(iterator))
```

Output

```text
1
2
StopIteration
```

---

## Iterator with a Loop

```python
numbers = [10, 20, 30]

iterator = iter(numbers)

for value in iterator:
    print(value)
```

Output

```text
10
20
30
```

## Real-World Example

Imagine a **TV remote**.

- The TV channels are the data.
- The remote acts as the iterator.
- Each press of the **Next** button returns one channel.
- When there are no more channels, the iteration stops.

---

## Iterable vs Iterator

| Iterable | Iterator |
|----------|----------|
| Stores a collection of data | Traverses the data one item at a time |
| Created using lists, tuples, strings, etc. | Created using `iter()` |
| Uses `iter()` to create an iterator | Uses `next()` to get the next element |

---

## Key Functions

| Function | Purpose |
|----------|---------|
| `iter()` | Creates an iterator from an iterable |
| `next()` | Returns the next element |

---

## Key Takeaways

- An iterator returns **one element at a time**.
- `iter()` converts an iterable into an iterator.
- `next()` retrieves the next value.
- `__iter__()` and `__next__()` implement the Iterator Protocol.
- When all elements are consumed, `StopIteration` is raised.
- Iterators are memory-efficient because they generate values on demand.
