# 🧬 Single & Multiple Inheritance in Python

## What is Inheritance?

**Inheritance** is one of the four pillars of Object-Oriented Programming (OOP). It allows a **child class** to inherit the properties (attributes) and behaviors (methods) of a **parent class**.

In simple words:

> **Inheritance allows one class to reuse the code of another class.**

---

# Why Use Inheritance?

- ♻️ Code Reusability
- 🚀 Avoid Code Duplication
- 🔧 Easy Maintenance
- 📈 Better Scalability
- 🧩 Supports Polymorphism

---

# Basic Terminology

| Term | Description |
|------|-------------|
| Parent Class | The class being inherited from. |
| Child Class | The class that inherits another class. |
| Base Class | Another name for Parent Class. |
| Derived Class | Another name for Child Class. |
| `super()` | Calls methods or constructors of the parent class. |
| Method Overriding | Redefining a parent's method inside the child class. |

---

# Types of Inheritance in Python

Python supports five types of inheritance:

1. Single Inheritance
2. Multiple Inheritance
3. Multilevel Inheritance
4. Hierarchical Inheritance
5. Hybrid Inheritance

# Single Inheritance

## Definition

**Single Inheritance** means one child class inherits from only one parent class.

```
Animal
   │
   ▼
 Dog
```

---

## Syntax

```python
class Parent:
    pass

class Child(Parent):
    pass
```

---

## Example

```python
class Animal:

    def __init__(self, name):
        self.name = name

    def eat(self):
        print(f"{self.name} is eating.")


class Dog(Animal):

    def bark(self):
        print(f"{self.name} is barking.")


dog = Dog("Rocky")

dog.eat()
dog.bark()
```

### Output

```text
Rocky is eating.
Rocky is barking.
```

---

## How it Works

When we create an object of `Dog`:

```python
dog = Dog("Rocky")
```

Python first searches inside the `Dog` class.

If it cannot find the requested method, it automatically searches inside the parent class (`Animal`).

Search Order:

```
Dog
 ↓
Animal
 ↓
object
```

---

# Method Overriding

A child class can replace the parent's implementation.

```python
class Animal:

    def speak(self):
        print("Animal Sound")


class Dog(Animal):

    def speak(self):
        print("Bark")


dog = Dog()
dog.speak()
```

### Output

```text
Bark
```

The child's method overrides the parent's method.

---

# Using super()

Instead of calling the parent class directly:

```python
Animal.__init__(self, name)
```

Python recommends using:

```python
super().__init__(name)
```

Example:

```python
class Animal:

    def __init__(self, name):
        self.name = name


class Dog(Animal):

    def __init__(self, name):
        super().__init__(name)
```

### Why use `super()`?

- Cleaner code
- Easier maintenance
- Supports Multiple Inheritance
- Recommended Python practice

# Multiple Inheritance

## Definition

**Multiple Inheritance** means one child class inherits from **more than one parent class**.

```
      Animal      Pet
         \        /
          \      /
           \    /
             Dog
```

The child class can use members from both parent classes.

---

## Syntax

```python
class Parent1:
    pass

class Parent2:
    pass

class Child(Parent1, Parent2):
    pass
```

---

## Example

```python
class Animal:

    def __init__(self, name):
        self.name = name


class Pet:

    def __init__(self, owner):
        self.owner = owner


class Dog(Animal, Pet):

    def __init__(self, owner, name):
        Animal.__init__(self, name)
        Pet.__init__(self, owner)

    def speak(self):
        print(f"{self.name} belongs to {self.owner} and is barking.")


dog = Dog("Hamza", "Rocky")

dog.speak()
```

### Output

```text
Rocky belongs to Hamza and is barking.
```

---

# Object Memory

```
Dog Object

name  ─────► Rocky
owner ─────► Hamza
```

`name` comes from `Animal`

`owner` comes from `Pet`

---

# Method Resolution Order (MRO)

Suppose both parent classes contain the same method.

```python
class Animal:

    def speak(self):
        print("Animal")


class Pet:

    def speak(self):
        print("Pet")


class Dog(Animal, Pet):
    pass


dog = Dog()

dog.speak()
```

### Output

```text
Animal
```

Why?

Because Python follows the **Method Resolution Order (MRO)**.

---

## Check the MRO

```python
print(Dog.mro())
```

Output

```text
[Dog, Animal, Pet, object]
```

Python searches methods in this order.

```
Dog
 ↓
Animal
 ↓
Pet
 ↓
object
```

---

# Diamond Problem

```
        A
      /   \
     B     C
      \   /
        D
```

Both `B` and `C` inherit from `A`.

`D` inherits from both `B` and `C`.

Now Python must decide:

> Which version of the method should be executed?

Python solves this using the **C3 Linearization Algorithm**, which creates a consistent Method Resolution Order (MRO).

# Using super() in Multiple Inheritance

Modern Python recommends cooperative inheritance.

```python
class Animal:

    def __init__(self, name, **kwargs):
        self.name = name
        super().__init__(**kwargs)


class Pet:

    def __init__(self, owner, **kwargs):
        self.owner = owner
        super().__init__(**kwargs)


class Dog(Animal, Pet):

    def __init__(self, name, owner):
        super().__init__(name=name, owner=owner)


dog = Dog("Rocky", "Hamza")

print(dog.name)
print(dog.owner)
```

This allows every parent class to initialize correctly.

# Single vs Multiple Inheritance

| Feature | Single | Multiple |
|----------|---------|----------|
| Number of Parents | One | Two or More |
| Complexity | Low | Higher |
| Readability | Easy | Moderate |
| MRO Required | No | Yes |
| Diamond Problem | No | Possible |
| Recommended | Yes | When Needed |

# Advantages of Inheritance

- Code Reusability
- Less Duplicate Code
- Easy Maintenance
- Better Project Structure
- Supports Polymorphism
- Easy to Extend Existing Code

---

# Disadvantages

- Tight Coupling
- Complex Hierarchies
- Harder Debugging
- Multiple Inheritance may create ambiguity

---

# Best Practices

- Use inheritance only when there is an **is-a** relationship.
- Prefer `super()` over direct parent class calls.
- Avoid unnecessary deep inheritance.
- Keep class hierarchies simple.
- Use Multiple Inheritance only when required.

````python
# Single Inheritance
# Parent Class 
class Car:
    def __init__(self,windows,doors,enginetype):
        self.window=windows
        self.doors=doors
        self.enginetype=enginetype

    def drive(self):
        print(f"It is an {self.enginetype} car")

car1=Car(4,4,"Petrol")
car1.drive()
````

````python
class Tesla(Car):
    def __init__(self,windows,doors,enginetype,is_selfdriving):
        super().__init__(windows,doors,enginetype)
        self.is_selfdriving=is_selfdriving
    def selfdrive(self):
        print(f"{self.is_selfdriving} Your Car is Self Driving Car")
    def detail(self):
        print(f"Your Car has {self.window} windows.{self.doors} Doors and is {self.enginetype} Car")

car2=Tesla(4,5,"Electric",True)
car2.selfdrive()
car2.detail()
````

````python
# using Mulitple Inheritance
class Animal:
    def __init__(self,name):
        self.name=name

class Pet:
    def __init__(self,owner):
        self.owner=owner


class Dog(Animal,Pet):
    def __init__(self,owner,name):
      Animal.__init__(self,name)
      Pet.__init__(self,owner)

    def speak(self):
        print(f" {self.name} is Barking whose owner name is  {self.owner.upper()}")

dog1=Dog("Hamza","Luffy")
dog1.speak()
````
