```python
### Encapsulation  with Getter and Setter MEthods
### Public,protected,private variables or access modifiers

class Person:
    def __init__(self,name,age):
        self.name=name    ## public variables
        self.age=age      ## public variables

def get_name(person):
    return person.name

person=Person("Hamza",32)
get_name(person)
```

```python
# Protected Variable
class person:
    def __init__(self,name,age):
        self._name=name
        self._age=age
class employee(person):
    def __init__(self, name, age):
        super().__init__(name, age)

def get_name(person):
    return f"you name is {person._name}"      

emp1=employee("Hamza",32)
# emp1._name
# emp1._age

# get_name(emp1)
p1=person("Arham",32)
p1._name
```

```python
class Person:
    def __init__(self,name,age):
        self.__name=name  ## Private access modifier or variable
        self.__age=age ## Private variable

    ## getter method for name
    def get_name(self):
        return self.__name
    
    ## setter method for name
    def set_name(self,name):
        self.__name=name

    # Getter method for age
    def get_age(self):
        return self.__age
    
    # Setter method for age
    def set_age(self, age):
        if age > 0:
            self.__age = age
        else:
            print("Age cannot be negative.")


person=Person("Hamza",22)

## Access and modify private variables using getter and setter

print(f"Your Name is {person.get_name()}")
print(f"Age is :{person.get_age()}")

person.set_age(35)
print(f"After callig Setter Method,Your Age is :{person.get_age()}")

person.set_age(-5)
    
```
