# 🔐 Encapsulation in Python

## What is Encapsulation?

**Encapsulation** is one of the four fundamental principles of Object-Oriented Programming (OOP).

It is the process of **wrapping data (variables) and methods (functions) into a single unit (class)** while controlling direct access to the data.

> **Definition:**  
> **Encapsulation is the process of bundling data and methods into a single class and restricting direct access to the data to protect it from unauthorized modification.**

---

# 🎯 Why Do We Use Encapsulation?

Encapsulation is used to:

- Protect sensitive data.
- Prevent unauthorized access.
- Improve data security.
- Improve code maintainability.
- Control how data is accessed and modified.

---

# 🏦 Real-Life Example

Imagine an **ATM Machine**.

You can:

- Deposit money ✅
- Withdraw money ✅
- Check balance ✅

But you **cannot directly change your account balance**.

❌ Wrong:

```
Balance = 1,000,000
```

Instead, you must use ATM functions like:

- Withdraw()
- Deposit()

The ATM controls access to your money.

This is **Encapsulation**.

---

# Without Encapsulation

```python
class BankAccount:

    def __init__(self):
        self.balance = 5000


account = BankAccount()

account.balance = 1000000

print(account.balance)
```

### Output

```text
1000000
```

Anyone can change the balance.

This is **not secure**.

---

# With Encapsulation

```python
class BankAccount:

    def __init__(self):
        self.__balance = 5000

    def deposit(self, amount):
        self.__balance += amount

    def get_balance(self):
        return self.__balance


account = BankAccount()

account.deposit(2000)

print(account.get_balance())
```

### Output

```text
7000
```

The balance is protected.

Users must use the provided methods.

# Access Modifiers Used in Encapsulation

## 1️⃣ Public Variable

Accessible from anywhere.

```python
self.name
```

---

## 2️⃣ Protected Variable

Intended for internal use.

```python
self._name
```

Python allows access, but it is only a convention.

---

## 3️⃣ Private Variable

Hidden using **Name Mangling**.

```python
self.__name
```

Direct access is not allowed.

---

# Getter Method

A **Getter** is used to read private data.

```python
def get_balance(self):
    return self.__balance
```

Usage

```python
print(account.get_balance())
```

---

# Setter Method

A **Setter** is used to update private data safely.

```python
def set_balance(self, amount):
    if amount >= 0:
        self.__balance = amount
```

Usage

```python
account.set_balance(10000)
```

# Complete Example

```python
class Student:

    def __init__(self, name, age):
        self.__name = name
        self.__age = age

    def get_name(self):
        return self.__name

    def set_name(self, name):
        self.__name = name


student = Student("Hamza", 22)

print(student.get_name())

student.set_name("Ali")

print(student.get_name())
```

### Output

```text
Hamza
Ali
```

````python
class Person:
    def __init__(self,name,age):
        self.__name=name  ## Private access modifier or variable
        self.__age=age ## Private variable

    ## getter method for name
    def get_name(self):
        return self.__name
    
    ## setter method for name
    def set_name(self,name):
        self.__name=name

    # Getter method for age
    def get_age(self):
        return self.__age
    
    # Setter method for age
    def set_age(self, age):
        if age > 0:
            self.__age = age
        else:
            print("Age cannot be negative.")


person=Person("Hamza",22)

## Access and modify private variables using getter and setter

print(f"Your Name is {person.get_name()}")
print(f"Age is :{person.get_age()}")

person.set_age(35)
print(f"After callig Setter Method,Your Age is :{person.get_age()}")

person.set_age(-5)
    
````
