# 🏛️ Classes & Objects in Python

## 📖 What is Object-Oriented Programming (OOP)?

Object-Oriented Programming (OOP) is a programming paradigm that organizes code using **classes** and **objects**. It helps build reusable, modular, and maintainable applications.

---

# 📌 What is a Class?

A **class** is a blueprint or template used to create objects.

It defines:

- Attributes (Variables)
- Methods (Functions)

### 🏠 Real-Life Example

Think of a **Car**.

A car has:

- Brand
- Color
- Model

And it can perform actions like:

- Start
- Stop
- Accelerate

Here, **Car** is the **Class**.

---

## 📝 Syntax

```python
class Student:
    pass
```

---

# 📌 What is an Object?

An **object** is an instance of a class.

A class is only a blueprint. The actual entity created from that blueprint is called an object.

### Example

```python
class Student:
    pass

student1 = Student()
student2 = Student()
```

Here:

- `Student` → Class
- `student1` → Object
- `student2` → Object

---

# 🔄 Class vs Object

| Class | Object |
|--------|--------|
| Blueprint | Real Instance |
| Defines structure | Stores actual data |
| Created once | Can create many objects |
| No data storage | Stores data |

---

# 🏗️ Constructor (`__init__()`)

A constructor is a special method that is automatically called whenever an object is created.

It is mainly used to initialize object data.

### Example

```python
class Student:

    def __init__(self, name, age):
        self.name = name
        self.age = age
```

---

# 🎯 Creating Objects

```python
class Student:

    def __init__(self, name, age):
        self.name = name
        self.age = age

student1 = Student("Hamza", 22)
student2 = Student("Ali", 20)
```

Each object stores its own data.

---

# 📦 Instance Variables

Instance variables belong to individual objects.

They are created using `self`.

```python
class Student:

    def __init__(self, name, age):
        self.name = name
        self.age = age
```

Here:

- `self.name`
- `self.age`

are **Instance Variables**.

Each object has its own separate copy.

---

# ⚙️ Instance Methods

Methods that operate on object data are called **Instance Methods**.

```python
class Student:

    def __init__(self, name):
        self.name = name

    def display(self):
        print(self.name)
```

Usage:

```python
student = Student("Hamza")
student.display()
```

**Output**

```text
Hamza
```

---

# 🔑 What is `self`?

`self` is a reference to the **current object (instance)**.

Python automatically passes the current object as the first argument to every instance method.

### Example

```python
class Student:

    def __init__(self, name):
        self.name = name

student = Student("Hamza")
```

Internally, Python executes:

```python
Student.__init__(student, "Hamza")
```

Therefore,

```python
self.name
```

actually means

```python
student.name
```

---

# ❓ Can We Change the Name of `self`?

Yes.

```python
class Student:

    def __init__(abc, name):
        abc.name = name
```

This code works.

> **Best Practice:** Always use `self` because it is the official Python convention.

```text

Ali
```

Each object stores independent data.

# 💻 Complete Example

```python
class Student:

    def __init__(self, name, age):
        self.name = name
        self.age = age

    def display(self):
        print(f"Name : {self.name}")
        print(f"Age  : {self.age}")

student1 = Student("Hamza", 22)
student2 = Student("Ali", 20)

student1.display()
student2.display()
```

**Output**

```text
Name : Hamza
Age  : 22

Name : Ali
Age  : 20
```

````python
# Creating Bank Account
class BankAccount:
    def __init__(self,owner,balance=0):
        self.name=owner
        self.balance=balance

    def deposit(self,amount):
        self.balance+=amount
        print(f"You Deposit {amount} in your account.Your Current Balance is {self.balance}")

    def withdraw(self,amount):
        self.balance-=amount
        print(f"You Withdraw {amount}.Your Current Balance is {self.balance}")

    def get_balance(self):
        print(f"Your Current Balance is {self.balance}")


acc1=BankAccount("Hamza",5000)
acc1.get_balance()   
acc1.deposit(300)
acc1.get_balance()
acc1.withdraw(1000)
acc1.get_balance()
````
