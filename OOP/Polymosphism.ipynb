# 🏛️ Abstract Class in Python

## What is an Abstract Class?

An **Abstract Class** is a special class that acts as a **blueprint** for other classes.

It contains:

- Normal methods ✅
- Abstract methods ✅ (methods without implementation)

Child classes **must implement** all abstract methods.

---

## Real-Life Example

Think of a school principal.

The principal says:

> "Every student must wear a uniform."

The principal only defines the **rule**.

Each student follows the rule in their own way.

Similarly, an Abstract Class only defines **what should happen**, not **how** it should happen.

---

## Example

```python
from abc import ABC, abstractmethod

class Animal(ABC):

    @abstractmethod
    def sound(self):
        pass
```

The `Animal` class says:

> Every animal must have a `sound()` method.

But it doesn't define the sound.

---

## Why can't we create an object of an Abstract Class?

Because an Abstract Class is **incomplete**.

Imagine buying a car that has:

- ✔ Wheels
- ✔ Doors
- ❌ No Engine

Can you drive it?

**No.**

Similarly, an Abstract Class has unfinished (abstract) methods, so Python doesn't allow creating its object.

```python
animal = Animal()
```

Output:

```text
TypeError:
Can't instantiate abstract class Animal...
```

Only a child class that implements all abstract methods can create objects.

```python
class Dog(Animal):

    def sound(self):
        print("Bark")

dog = Dog()
dog.sound()
```

**Output**

```text
Bark
```

---

## Key Points

- Abstract Classes act as blueprints.
- They may contain abstract methods.
- Child classes must implement abstract methods.
- Objects cannot be created because the class is incomplete.

````python
# Base class 
class Animal:
    def speak(self):
        print("Animalis  barking")

# Derived Class 
class Dog(Animal):
    def speak(self):
        return f"wooof!"
class Cat(Animal):
    def speak(self):
        return f"Meow!"
    pass


def speak(Animal):
    print( Animal.speak())

dog1=Dog()
cat1=Cat() 
speak(cat1)
speak(dog1)
````

````python
# Calculating Area of Rectangle Circle Using Functions and Methods
# from Math import pi
class Shape:
    def area(self):
        print("The Area of the Figure")
class Rectangle(Shape):
    def __init__(self,width,height):
        self.width=width
        self.height=height

    def area(self):
        return self.width*self.height  
class Circle(Shape):
    def __init__(self,radius):
        self.radius=radius

    def area(self):
        return  3.14*self.radius*self.radius
    


def area_shape(Shape):
    return f"Area is {Shape.area()}"


    
c1=Circle(3)
r1=Rectangle(3,4)
print(f"{area_shape(c1)}")
print(f"{area_shape(r1)}")

                   
````

````python
from abc import ABC,abstractmethod

## Define an abstract class
class Vehicle(ABC):
    @abstractmethod
    def start_engine(self):
        pass

## Derived class 1
class Car(Vehicle):
    def start_engine(self):
        return "Car engine started"
    
## Derived class 2
class Motorcycle(Vehicle):
    def start_engine(self):
        return "Motorcycle enginer started"
    
# Function that demonstrates polymorphism
def start_vehicle(vehicle):
    print(vehicle.start_engine())

## create objects of cAr and Motorcycle

car = Car()
motorcycle = Motorcycle()

start_vehicle(car)
````
