```python
# Base class 
class Animal:
    def speak(self):
        print("Animalis  barking")

# Derived Class 
class Dog(Animal):
    def speak(self):
        return f"wooof!"
class Cat(Animal):
    def speak(self):
        return f"Meow!"
    pass


def speak(Animal):
    print( Animal.speak())

dog1=Dog()
cat1=Cat() 
speak(cat1)
speak(dog1)
```

```python
# Calculating Area of Rectangle Circle Using Functions and Methods
# from Math import pi
class Shape:
    def area(self):
        print("The Area of the Figure")
class Rectangle(Shape):
    def __init__(self,width,height):
        self.width=width
        self.height=height

    def area(self):
        return self.width*self.height  
class Circle(Shape):
    def __init__(self,radius):
        self.radius=radius

    def area(self):
        return  3.14*self.radius*self.radius
    


def area_shape(Shape):
    return f"Area is {Shape.area()}"


    
c1=Circle(3)
r1=Rectangle(3,4)
print(f"{area_shape(c1)}")
print(f"{area_shape(r1)}")

                   
```
