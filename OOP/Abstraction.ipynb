# Python Example

```python
from abc import ABC, abstractmethod

class Animal(ABC):

    @abstractmethod
    def sound(self):
        pass


class Dog(Animal):

    def sound(self):
        print("Bark")


dog = Dog()
dog.sound()
```

### Output

```text
Bark
```

The `Animal` class only defines **what** should happen (`sound()`), while the `Dog` class defines **how** it happens.

# 🎭 Abstraction in Python

## What is Abstraction?

**Abstraction** is one of the four fundamental principles of Object-Oriented Programming (OOP).

It means:

> **Hiding the implementation details (How) and showing only the essential functionality (What).**

The user only knows **what** a program can do, not **how** it does it internally.

---

# 🎯 Why Do We Use Abstraction?

We use Abstraction to:

- Hide unnecessary implementation details.
- Reduce code complexity.
- Improve code readability.
- Make programs easier to use.
- Focus only on essential functionality.

---

# 🚗 Real-Life Example

Think about a **Car**.

To start the car, you simply press the **Start** button.

You don't need to know:

- How the battery works.
- How fuel is injected.
- How the engine starts.

You only know:

```text
Press Start Button
```

The internal process is hidden from you.

# 💻 Coding Example

```python
class Car:

    def start(self):
        self.__check_battery()
        self.__inject_fuel()
        self.__start_engine()
        print("Car Started")

    def __check_battery(self):
        print("Checking Battery...")

    def __inject_fuel(self):
        print("Injecting Fuel...")

    def __start_engine(self):
        print("Starting Engine...")
```

Now create an object.

```python
car = Car()
car.start()
```

### Output

```text
Checking Battery...
Injecting Fuel...
Starting Engine...
Car Started
```

# Where is the Abstraction?

The user only writes:

```python
car.start()
```

The user **does not know**:

- How the battery is checked.
- How fuel is injected.
- How the engine starts.

The implementation is hidden inside the class.

The user only sees the **start()** method.

This is **Abstraction**.

---

# How is Abstraction Achieved in Python?

Python commonly achieves Abstraction using:

- `ABC` (Abstract Base Class)
- `@abstractmethod`

---

# Abstract Class Example

```python
from abc import ABC, abstractmethod

class Animal(ABC):

    @abstractmethod
    def sound(self):
        pass
```

The `Animal` class only says:

> Every Animal must have a **sound()** method.

It does **not** define how the sound is produced.

---

# Child Classes

```python
class Dog(Animal):

    def sound(self):
        print("Bark")
```

```python
class Cat(Animal):

    def sound(self):
        print("Meow")
```

Now create objects.

```python
dog = Dog()
cat = Cat()

dog.sound()
cat.sound()
```

### Output

```text
Bark
Meow
```

---

# What is Hidden Here?

The `Animal` class only exposes:

```python
sound()
```

But the actual implementation is hidden inside each child class.

For example:

Dog knows:

```python
print("Bark")
```

Cat knows:

```python
print("Meow")
```

The parent class doesn't know **how** each animal makes its sound.

It only defines **what** should happen.

This is also an example of **Abstraction**.

---

# Difference Between Abstraction and Abstract Class

| Abstraction | Abstract Class |
|-------------|----------------|
| A concept. | A tool to implement abstraction. |
| Hides implementation details. | Provides a blueprint for child classes. |
| Focuses on **What**. | Defines common rules for child classes. |

````python
class Car:

    def start(self):
        self.__check_battery()
        self.__inject_fuel()
        self.__start_engine()
        print("Car Started")

    def __check_battery(self):
        print("Checking Battery...")

    def __inject_fuel(self):
        print("Injecting Fuel...")

    def __start_engine(self):
        print("Starting Engine...")
````

````python
car=Car()
car.start()
````

````python
from abc import ABC,abstractmethod
## Abstract base cclass
class Vehicle(ABC):
    def drive(self):
        print("The vehicle is used for driving")

    @abstractmethod
    def start_engine(self):
        pass

class Car(Vehicle):
    def start_engine(self):
        print("Car enginer started")

def operate_vehicle(vehicle):
    vehicle.start_engine()
    vehicle.drive()

car=Car()
operate_vehicle(car)
````
