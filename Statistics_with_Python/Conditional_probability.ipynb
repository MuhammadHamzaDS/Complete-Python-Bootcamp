# Conditional Probability

Conditional probability is the probability of an event occurring when another event has already occurred.

## Formula

$$
P(A \mid B) = \frac{P(A \cap B)}{P(B)}
$$

Where:

* $P(A \mid B)$ = Probability of event A when event B has already occurred
* $P(A \cap B)$ = Probability that both A and B occur
* $P(B)$ = Probability of event B

## Example

A bag contains `6` red balls. Among them, `2` balls are striped.

The probability that a ball is striped, given that it is red, is:

$$
P(\text{Striped} \mid \text{Red}) = \frac{2}{6}
$$

$$
P(\text{Striped} \mid \text{Red}) = 0.3333 = 33.33%
$$

## Python Implementation

```python
def conditional_probability(matching_cases, conditioned_cases):
    if conditioned_cases == 0:
        raise ValueError("Conditioned cases cannot be zero.")

    return matching_cases / conditioned_cases


red_and_striped_balls = 2
total_red_balls = 6

result = conditional_probability(
    red_and_striped_balls,
    total_red_balls
)

print("Probability:", result)
print("Percentage:", round(result * 100, 2), "%")
```

## Key Point

When a condition is given, the sample space changes.

In this example, the condition is that the ball is red, so only the `6` red balls are considered.


````python
def conditional_probability(red_and_striped_balls, total_red_balls):
    if total_red_balls == 0:
        raise ValueError("Total red balls cannot be zero.")

    probability = red_and_striped_balls / total_red_balls
    return probability


red_and_striped_balls = int(input("Enter Red_Striped_Balls:"))
total_red_balls = int(input("Enter Total Numbers of Red Balls:"))

striped_Chance = conditional_probability(
    red_and_striped_balls,
    total_red_balls
)

print("IF the Ball is Already Red There Striped Chances are:", striped_Chance)
print("Percentage:", striped_Chance* 100, "%")
````
