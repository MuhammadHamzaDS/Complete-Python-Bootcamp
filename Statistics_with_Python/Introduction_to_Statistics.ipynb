# Introduction to Statistics

## What is Statistics?

Statistics is the process of collecting, organizing, analyzing, interpreting, and presenting data.

It helps us convert raw data into useful information and make better decisions.

---

## Why Do We Use Statistics?

Statistics is used to:

* Summarize large datasets
* Understand patterns and trends
* Compare different groups
* Make predictions
* Support decision-making
* Measure uncertainty

---

## Real-Life Applications

Statistics is used in many fields:

* Schools analyze student performance
* Businesses analyze sales and customers
* Hospitals study patient data
* Governments conduct population surveys
* Data scientists train machine learning models

---

## Main Types of Statistics

Statistics has two main branches:

### 1. Descriptive Statistics

Descriptive statistics summarizes and describes the available data.

Common methods include:

* Mean
* Median
* Mode
* Range
* Variance
* Standard deviation
* Tables and charts

### Example

Suppose the marks of five students are:

```text
70, 80, 75, 90, 85
```

Descriptive statistics can calculate the average, middle, lowest, and highest marks.

---

### 2. Inferential Statistics

Inferential statistics uses sample data to make conclusions or predictions about a larger population.

Common methods include:

* Sampling
* Confidence intervals
* Hypothesis testing
* Regression
* Prediction

### Example

A university selects `200` students and uses their responses to understand the opinions of all university students.

---

## Descriptive vs Inferential Statistics

| Descriptive Statistics            | Inferential Statistics                     |
| --------------------------------- | ------------------------------------------ |
| Describes available data          | Draws conclusions about a population       |
| Uses averages, tables, and charts | Uses samples and probability               |
| Summarizes past or current data   | Makes estimates and predictions            |
| Example: Average class marks      | Example: Predicting university performance |

---

## Basic Statistical Terms

### Data

Data consists of collected facts, numbers, or observations.

Example:

```text
Student names, marks, ages, and attendance
```

### Dataset

A dataset is an organized collection of related data.

### Population

A population is the complete group that we want to study.

Example:

```text
All students in a university
```

### Sample

A sample is a smaller group selected from the population.

Example:

```text
200 selected university students
```

### Variable

A variable is a characteristic that can have different values.

Examples:

```text
Age, marks, height, and salary
```

### Observation

An observation is one complete record in a dataset.

For example, the information of one student represents one observation.

---

## Key Takeaway

Statistics helps us understand data and make informed decisions.

The two main branches are:

1. **Descriptive Statistics** — summarizes available data
2. **Inferential Statistics** — draws conclusions using sample data


````python
import statistics

student_marks = [70, 80, 75, 90, 85]

mean_marks = statistics.mean(student_marks)
median_marks = statistics.median(student_marks) # it will Automatically Sort the Data 
minimum_marks = min(student_marks)
maximum_marks = max(student_marks)

print("Mean:", mean_marks)
print("Median:", median_marks)
print("Minimum:", minimum_marks)
print("Maximum:", maximum_marks)
````

## Understanding the Result

* **Mean = 80:** The average marks are `80`.
* **Median = 80:** The middle value after sorting the data is `80`.
* **Minimum = 70:** The lowest marks are `70`.
* **Maximum = 90:** The highest marks are `90`.
