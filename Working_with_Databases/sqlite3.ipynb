# SQLite3 (Python)

## Definition
**SQLite3** is a lightweight, serverless, file-based relational database built into Python.

---

## Import

```python
import sqlite3
```

---

## Connect Database

```python
connection = sqlite3.connect("company.db")
```

> Creates or connects to a database.

---

## Create Cursor

```python
cursor = connection.cursor()
```

> Executes SQL queries.

---

## Execute Query

```python
cursor.execute("SELECT * FROM employees")
```

> Executes a single SQL statement.

---

## Insert Multiple Records

```python
cursor.executemany(query, data)
```

> Executes the same query for multiple records.

---

## Save Changes

```python
connection.commit()
```

> Permanently saves changes.

---

## Fetch Data

```python
cursor.fetchone()   # One row
cursor.fetchmany(3) # Multiple rows
cursor.fetchall()   # All rows
```

---

## Close Connection

```python
connection.close()
```

> Closes the database connection.

---

## Workflow

```text
Connect → Cursor → Execute → Fetch → Commit → Close
```

---

## Key Methods

| Method | Purpose |
|---------|---------|
| `connect()` | Connect database |
| `cursor()` | Create cursor |
| `execute()` | Execute one query |
| `executemany()` | Execute multiple queries |
| `fetchall()` | Get all rows |
| `commit()` | Save changes |
| `close()` | Close connection |

---

## Interview Definitions

- **SQLite3:** A lightweight, serverless, file-based relational database.
- **Connection:** Connects Python to the database.
- **Cursor:** Executes SQL queries and fetches results.
- **commit():** Saves changes permanently.
- **close():** Closes the database connection.

---

## Memory Trick

```text
connect()  → Open
cursor()   → Query
execute()  → Run
fetchall() → Read
commit()   → Save
close()    → Exit
```

````python
connection=sqlite3.connect("Sales.db")
cursor=connection.cursor()
````

````python
cursor.execute('''
Create Table if Not exists Sales(
               id Integer Primary key,
               date text Not Null,
               product  Text Not Null,
               sale integer,
               region text
               )

''')
sale_data=[
    ('2023-01-01', 'Product1', 100, 'North'),
    ('2023-01-02', 'Product2', 200, 'South'),
    ('2023-01-03', 'Product1', 150, 'East'),
    ('2023-01-04', 'Product3', 250, 'West'),
    ('2023-01-05', 'Product2', 300, 'North')
]
cursor.executemany('''
Insert Into Sales(data,product,sale,region)
               values(?,?,?,?)

''',sale_data)
    
connection.commit()
````

````python
#  Query data from the sales table
cursor.execute('SELECT * FROM sales')
rows = cursor.fetchall()

# Print the queried data
for row in rows:
    print(row)
````

````python
cursor.execute('''


Delete from Sales
where sale>100
                         
''')
````

````python
#  Query data from the sales table
cursor.execute('SELECT * FROM sales')
rows = cursor.fetchall()

# Print the queried data
for row in rows:
    print(row)

connection.commit()
````
