```python
from faker import Faker
import sqlite3
import random

fake = Faker()

pakistan_cities = [
    "Lahore",
    "Karachi",
    "Islamabad",
    "Rawalpindi",
    "Faisalabad",
    "Multan",
    "Sahiwal",
    "Peshawar",
    "Quetta",
    "Gujranwala"
]

# Connect to SQLite database
connection = sqlite3.connect("students.db")
cursor = connection.cursor()

# Create students table
cursor.execute("""
CREATE TABLE IF NOT EXISTS students (
    id INTEGER PRIMARY KEY AUTOINCREMENT,
    name TEXT NOT NULL,
    email TEXT NOT NULL,
    age INTEGER,
    marks INTEGER,
    city TEXT
)
""")

# Generate and insert 100 records
for _ in range(100):
    student_data = (
        fake.name(),
        fake.email(),
        random.randint(18, 25),
        random.randint(40, 100),
        random.choice(pakistan_cities)
    )

    cursor.execute("""
    INSERT INTO students (name, email, age, marks, city)
    VALUES (?, ?, ?, ?, ?)
    """, student_data)

# Save changes and close connection
connection.commit()
connection.close()

print("Fake data generated and stored successfully.")
```
