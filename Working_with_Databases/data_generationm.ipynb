```python
from faker import Faker
import random
import csv

fake=Faker()
with open('student.csv','w') as file:
    writer=csv.writer(file)


    writer.writerow(["name","Email","Age","Marks","City"])
    for i in range(100):
     writer.writerow([
        fake.name(),
        fake.email(),
        random.randint(18,25),
        random.randint(40,100),
        fake.city()

    ])


print("Fake Data is Generated Successfully")
```
