# 📁 Python File Handling

## 📖 Definition

> **File Handling** is the process of creating, reading, writing, updating, and managing files using Python.

---

# 📂 Types of Files

### 1. Text Files

Human-readable files.

**Examples**

- `.txt`
- `.csv`
- `.json`
- `.py`

---

### 2. Binary Files

Store data as bytes and are not human-readable.

**Examples**

- `.jpg`
- `.png`
- `.mp3`
- `.mp4`
- `.pdf`
- `.docx`

---

# 📌 open()

Used to open a file.

```python
file = open("example.txt", "r")
```

---

# 📌 File Modes

| Mode | Description |
|------|-------------|
| `r` | Read |
| `w` | Write (Overwrite) |
| `a` | Append |
| `x` | Create |
| `r+` | Read & Write |
| `w+` | Write & Read |
| `a+` | Append & Read |
| `rb` | Read Binary |
| `wb` | Write Binary |
| `ab` | Append Binary |

---

# 📌 close()

Closes the opened file.

```python
file.close()
```

---

# 📌 with open()

Recommended way to work with files.

```python
with open("example.txt", "r") as file:
    print(file.read())
```

### ✅ Advantages

- Automatically closes the file
- Cleaner code
- Prevents resource leaks

---

# 📌 Reading a File

### Read Entire File

```python
file.read()
```

### Read Specific Characters

```python
file.read(10)
```

### Read One Line

```python
file.readline()
```

### Read All Lines

```python
file.readlines()
```

### Read Using Loop

```python
for line in file:
    print(line)
```

---

# 📌 Writing to a File

```python
with open("example.txt", "w") as file:
    file.write("Hello World")
```

> **Note:** `w` removes previous content.

---

# 📌 Append Data

```python
with open("example.txt", "a") as file:
    file.write("\nPython")
```

Adds data without removing previous content.

---

# 📌 writelines()

Writes multiple strings.

```python
lines = [
    "Python\n",
    "Java\n",
    "C++\n"
]

with open("example.txt", "w") as file:
    file.writelines(lines)
```

---

# 📌 seek()

Moves the file cursor to a specific position.

```python
file.seek(0)
```

### Common Uses

- Read file again
- Move to a specific position
- Read/write from a desired location

---

# 📌 tell()

Returns the current cursor position.

```python
print(file.tell())
```

---

# 📌 Binary Files

### Read Binary

```python
with open("image.jpg", "rb") as file:
    data = file.read()
```

### Write Binary

```python
with open("copy.jpg", "wb") as file:
    file.write(data)
```

### Append Binary

```python
with open("file.bin", "ab") as file:
    file.write(b"Hello")
```

---

# 📌 Best Practices

- ✅ Use `with open()`
- ✅ Always close files
- ✅ Use correct file mode
- ✅ Use `seek()` to reposition the cursor
- ✅ Use binary modes (`rb`, `wb`, `ab`) for binary files

````python
#Read a text file and count the number of lines, words, and characters.
# Counting lines, words, and characters in a text file
def count_text_file(file_path):
    with open(file_path, 'r') as file:
        lines = file.readlines()
        line_count = len(lines)
        word_count = sum(len(line.split()) for line in lines)
        char_count = sum(len(line) for line in lines)
    return line_count, word_count, char_count

file_path = 'example.txt'
lines, words, characters = count_text_file(file_path)
print(f'Lines: {lines}, Words: {words}, Characters: {characters}')
````
