### Introduction To Tuples:
- A tuple is an ordered, immutable sequence of elements that can contain heterogeneous data types. Once created, its contents cannot be modified, making it suitable for fixed collections of data.
#### Key Characteristics
- Ordered → Maintains insertion order
- Immutable → Cannot be changed after creation
- Heterogeneous → Can store multiple data types
- Indexed → Elements accessed via index (starting from 0)
- Allows duplicates → Same values can appear multiple times

```python
## creating a tuple
empty_tuple=()
print(empty_tuple)
print(type(empty_tuple))
```

```python
lst=list()
print(type(lst))
tpl=tuple()
print(type(tpl))
```

```python
numbers=tuple([1,2,3,4,5,6])
numbers
list((1,2,3,4,5,6))
```

```python
mixed_tuple=(1,"Hello World",3.14, True)
print(mixed_tuple)
```

```python
print(numbers[2])
print(numbers[-1])
```

```python
numbers[0:4]
```

```python
## Tuple Operations

concatenation_tuple=numbers + mixed_tuple
print(concatenation_tuple)
```

```python
mixed_tuple * 3
```

```python
numbers *3
```

```python
## Immutable Nature Of Tuples
## Tuples are immutable, meaning their elements cannot be changed once assigned.

lst=[1,2,3,4,5]
print(lst)

lst[1]="Krish"
print(lst)
```

```python
## Tuple Methods
print(numbers.count(1))
print(numbers.index(3))
```

```python
## Packing and Unpacking tuple
## packing
packed_tuple=1,"Hello",3.14
print(packed_tuple)
```

```python
##unpacking a tuple
a,b,c=packed_tuple

print(a)
print(b)
print(c)
```

```python
## Unpacking with *
numbers=(1,2,3,4,5,6)
first,*middle,last=numbers
print(first)
print(middle)
print(last)
```

```python
## Nested Tuple
## Nested List
lst=[[1,2,3,4],[6,7,8,9],[1,"Hello",3.14,"c"]]
lst[0][0:3]
```

```python
lst=[[1,2,3,4],[6,7,8,9],(1,"Hello",3.14,"c")]
lst[2][0:3]
```

```python
nested_tuple = ((1, 2, 3), ("a", "b", "c"), (True, False))

## access the elements inside a tuple
print(nested_tuple[0])
print(nested_tuple[1][2])   
```

```python
## iterating over nested tuples
for sub_tuple in nested_tuple:
    for item in sub_tuple:
        print(item,end=" ")
    print()
```

### Conclusion

They are commonly used in data structures, function arguments and return values, and as dictionary keys.
