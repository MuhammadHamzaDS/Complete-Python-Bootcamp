# Python Logging

## What is Logging?

**Logging** is used to record events, errors, warnings, and debugging information while a program is running. It helps developers monitor, debug, and maintain applications.

---

# Configure Logging

```python
import logging

logging.basicConfig(
    filename="app.log",
    filemode="w",
    level=logging.DEBUG,
    format="%(asctime)s - %(name)s - %(levelname)s - %(message)s",
    datefmt="%Y-%m-%d %H:%M:%S"
)
```

---

# Log Levels

| Level | Function | Purpose |
|--------|----------|---------|
| DEBUG | `logging.debug()` | Detailed debugging information |
| INFO | `logging.info()` | General program information |
| WARNING | `logging.warning()` | Indicates a possible problem |
| ERROR | `logging.error()` | Records an error |
| CRITICAL | `logging.critical()` | Records a critical/fatal error |

---

# Example

```python
import logging

logging.basicConfig(
    filename="app.log",
    level=logging.DEBUG,
    format="%(asctime)s - %(levelname)s - %(message)s"
)

logging.debug("Debug Message")
logging.info("Info Message")
logging.warning("Warning Message")
logging.error("Error Message")
logging.critical("Critical Message")
```

---

# Common Parameters

| Parameter | Description |
|-----------|-------------|
| `filename` | Log file name |
| `filemode` | File mode (`w` = overwrite, `a` = append) |
| `level` | Minimum log level to record |
| `format` | Format of each log message |
| `datefmt` | Date and time format |

---

# Common Format Specifiers

| Specifier | Description |
|-----------|-------------|
| `%(asctime)s` | Date & Time |
| `%(name)s` | Logger name |
| `%(levelname)s` | Log level name |
| `%(levelno)s` | Log level number |
| `%(message)s` | Log message |
| `%(filename)s` | File name |
| `%(funcName)s` | Function name |
| `%(lineno)d` | Line number |

# Logging Inside Functions

```python
from logger import logging

def add(a, b):
    logging.debug("Addition function started")
    result = a + b
    logging.info(f"Result = {result}")
    return result

logging.info("Program Started")
add(2, 3)
```

---

# Common Mistake

❌ Incorrect

```python
logging.DEBUG("Message")
```

✅ Correct

```python
logging.debug("Message")
```

**Reason:**

- `logging.DEBUG` → Constant (`10`)
- `logging.debug()` → Function that writes a log message.

---

# Best Practices

- Use **DEBUG** for debugging.
- Use **INFO** for normal execution messages.
- Use **WARNING** for recoverable issues.
- Use **ERROR** for failures.
- Use **CRITICAL** for fatal errors.
- Store logs in a separate `.log` file.
- Use meaningful log messages.

---

# Key Takeaways

- Logging helps monitor and debug applications.
- Configure logging once using `logging.basicConfig()`.
- Choose the correct log level based on message importance.
- Use `logging.debug()`, `logging.info()`, etc., not `logging.DEBUG()`.
- Log files make debugging much easier in real-world projects.
