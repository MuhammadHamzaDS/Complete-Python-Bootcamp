# Variable-Length Arguments

## Definition

**Variable-length arguments** allow a function to accept a **variable (unknown) number of arguments** at runtime, making the function more flexible and reusable.

### Types

- **`*args`** → Accepts multiple **positional arguments** and stores them in a **tuple**.
- **`**kwargs`** → Accepts multiple **keyword arguments** and stores them in a **dictionary**.

## Interview Definition

> **Variable-length arguments allow a function to accept an unspecified number of arguments. In Python, `*args` collects positional arguments into a tuple, while `**kwargs` collects keyword arguments into a dictionary.**

```python
# *args
def add(*numbers):
    return sum(numbers)

print(add(10, 20))
print(add(10, 20, 30))
```

```python
# *kwargs
def student(**kwargs):
    print(kwargs)

student(name="Hamza", age=22, course="Python")
def skills(**kwargs):
    for key,values in kwargs.items():
        print(f"{key}:Value")

skills(name="Khan",age=33,city="sahiwal")       
```

```python
# Map Function with Multiple Iterables
numbers1=[1,2,3,4]
numbers2=[3,4,5,6]
added_numbers=list(map(lambda x,y:x+y,numbers1,numbers2 ))
print(added_numbers)
```

```python
# Map with Single Iterbale

def sqaure(a):
    return a*a
nums=[2,3,4,5,6,7]
sqaure_num=list(map(sqaure,nums))
print(sqaure_num)
```

```python
# map() Function to Convert String to integers
numbers=['1','2','3','4','5']
to_integer=list(map(int,numbers))
print(to_integer)
```

```python
# Conert Lowercase to Uppercase
words=['apple','banana','grapes']
uppercase=list(map(str.upper,words))
print(uppercase)
```
