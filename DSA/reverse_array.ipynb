# Reverse a List in Python

Reversing a list means arranging its elements in the opposite order.

### Example

```python
Input  : [10, 20, 30, 40, 50]
Output : [50, 40, 30, 20, 10]
```

---

## Method 1 — `reverse()`

The `reverse()` method reverses the original list **in place**, meaning the original list is modified.

```python
numbers = [10, 20, 30, 40, 50]

numbers.reverse()

print(numbers)
```

---

## Method 2 — Slicing (`[::-1]`)

Slicing creates a **new reversed copy** of the list while leaving the original list unchanged.

```python
numbers = [10, 20, 30, 40, 50]

reversed_list = numbers[::-1]

print(reversed_list)
```

---

## Method 3 — `reversed()`

The `reversed()` function returns a reverse iterator. Convert it into a list using `list()`.

```python
numbers = [10, 20, 30, 40, 50]

reversed_list = list(reversed(numbers))

print(reversed_list)
```

---

## Method 4 — Using a Loop

This method manually traverses the list from the last index to the first index.

```python
numbers = [10, 20, 30, 40, 50]

reversed_list = []

for i in range(len(numbers)-1, -1, -1):
    reversed_list.append(numbers[i])

print(reversed_list)
```

---

## Comparison

| Method | Modifies Original List | Extra Space | Time Complexity |
|---------|:----------------------:|:-----------:|:---------------:|
| `reverse()` | ✅ Yes | O(1) | O(n) |
| `[::-1]` | ❌ No | O(n) | O(n) |
| `reversed()` | ❌ No | O(n) | O(n) |
| Loop | ❌ No | O(n) | O(n) |

---

## Key Takeaways

- Use **`reverse()`** to modify the original list.
- Use **`[::-1]`** to create a new reversed list.
- Use **`reversed()`** when an iterator is required.
- Use the **loop approach** to understand indexing and backward traversal.

````python

# Original List
numbers = [10, 20, 30, 40, 50]

print("Before Reverse:", numbers)

# Reverse the original list
numbers.reverse()

print("After Reverse :", numbers)
````

````python
# Original List
numbers = [10, 20, 30, 40, 50]

# Create a new reversed list
reversed_list = numbers[::-1]

print("Original List :", numbers)
print("Reversed List:", reversed_list)
````

````python
# Original List
numbers = [10, 20, 30, 40, 50]

# Reverse using reversed()
reversed_list = list(reversed(numbers))

print("Original List :", numbers)
print("Reversed List:", reversed_list)
````

````python
# Original List
numbers = [10, 20, 30, 40, 50]

# Empty list to store reversed elements
reversed_list = []

# Traverse from last index to first
for i in range(len(numbers) - 1, -1, -1):
    reversed_list.append(numbers[i])

print("Original List :", numbers)
print("Reversed List:", reversed_list)
````
