# Kadane's Algorithm

## 📌 What is Kadane's Algorithm?

**Kadane's Algorithm** is an efficient algorithm used to find the **maximum sum of a contiguous (continuous) subarray** within an array.

> **Simple Definition:**  
> It finds the consecutive group of numbers whose total sum is the largest.

---

# 🧒 Think Like a Kid (Real-Life Example)

Imagine you're playing a game where every day you either **earn** or **lose** coins.

| Day | Coins |
|-----|------:|
| 1 | -2 |
| 2 | 3 |
| 3 | -1 |
| 4 | 4 |
| 5 | -5 |
| 6 | 2 |

### Meaning

- Positive number (**+**) → You earned coins 💰
- Negative number (**-**) → You lost coins 😔

Your goal is:

> **Find the continuous days where you earned the most coins overall.**

---

## 🤔 What is the Best Choice?

Possible continuous groups:

```text
[-2] = -2

[3] = 3

[3, -1] = 2

[3, -1, 4] = 6 ✅

[4] = 4

[2] = 2
```

The maximum sum is:

```text
[3, -1, 4] = 6
```

So the answer is:

```text
Maximum Sum = 6
```

---

# 🧠 The Main Idea

At every number, ask yourself only **one question**:

> **Should I continue with my current group, or should I start a new group from this number?**

That's the entire idea behind Kadane's Algorithm.

# 💻 Python Code

```python
def kadane_algorithm(arr):
    current_sum = arr[0]
    maximum_sum = arr[0]

    for num in arr[1:]:

        # Continue previous group
        # OR
        # Start a new group

        current_sum = max(num, current_sum + num)

        # Update maximum sum if needed
        maximum_sum = max(maximum_sum, current_sum)

    return maximum_sum


numbers = [-2, 3, -1, 4, -5, 2]

answer = kadane_algorithm(numbers)

print("Maximum Subarray Sum:", answer)
```

### Output

```text
Maximum Subarray Sum: 6
```

# 📊 Time Complexity

```text
O(n)
```

Because we visit each element **only once**.

---

# 💾 Space Complexity

```text
O(1)
```

Because we only use two variables:

- `current_sum`
- `maximum_sum`

No extra array or list is created.

---

# 📚 Interview Definition

> **Kadane's Algorithm is an efficient algorithm used to find the maximum sum of a contiguous subarray in a one-dimensional array. It solves the problem in linear time O(n) by deciding at each element whether to extend the current subarray or start a new one.**

---

# 📝 Easy Formula to Remember

current_sum = max(current_number,
                  current_sum + current_number)

# ⭐ Memory Trick

Imagine you're carrying a **bag of money**.

Every new number represents today's earnings or loss.

Ask yourself:

> **Is my current bag still worth carrying?**

- ✅ **Yes** → Add today's money to the bag.
- ❌ **No** → Throw the old bag away and start a new one with today's money.

This simple decision at every step is exactly how **Kadane's Algorithm** works.

````python
# With Table Formatting
#For Advanced Programmers

def kadane_algorithm(arr):
    # Initialize both variables with the first element
    current_sum = arr[0]
    maximum_sum = arr[0]

    print(f"{'Number':<8}{'Current Sum':<15}{'Maximum Sum'}")
    print("-" * 35)

    print(f"{arr[0]:<8}{current_sum:<15}{maximum_sum}")

    # Traverse the remaining elements
    for num in arr[1:]:

        # Decide whether to continue the current subarray
        # or start a new subarray from the current number
        current_sum = max(num, current_sum + num)

        # Update the maximum sum found so far
        maximum_sum = max(maximum_sum, current_sum)

        print(f"{num:<8}{current_sum:<15}{maximum_sum}")

    return maximum_sum


# Driver Code
numbers = [-2, 3, -1, 4, -5, 2]

print("Input Array:", numbers)
print()

answer = kadane_algorithm(numbers)

print("\nMaximum Subarray Sum:", answer)
````

````python
# Simple Clear Concept For Beginners
def kadane_algorithm(numbers):
    current_sum = numbers[0]
    maximum_sum = numbers[0]

    for number in numbers[1:]:
        # Purane group ke saath rahen ya fresh start karein
        current_sum = max(number, current_sum + number)

        # Ab tak ka sab se bada sum save karein
        maximum_sum = max(maximum_sum, current_sum)

    return maximum_sum


numbers = [-2, 3, -1, 4, -5, 2]

result = kadane_algorithm(numbers)

print("Maximum Subarray Sum:", result)
````
